Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím \
Ví dụ: Ta nhập h = 4:  
<pre>
   *
  * *      
 *   *
*******

```python
h = int(input("nhập độ cao tam giác: "))

for i in range (1, h+1):
  if i==1: print(" "*(h-1) + "*")
  elif 1<i<h: print(" "*(h-i) + "*" + " "*(2*i-3) + "*" )
  elif i==h: print("*"*(2*i-1))

'''
khoảng giữa: bắt đầu từ dòng i=2 làm mốc
có tổng cộng i-2 dòng giữa, xuống 1 dòng sẽ có thêm 2 khoảng
=>2(i-2)
cộng thêm mốc đầu tiên
=> 1 + 2(i-2) = 2i-3
'''
```

Bài 9: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số chính phương hay
không? Số chính phương là số có căn bậc hai là một số nguyên VD: 1, 4, 9, 16, 25….

```python
a = int(input("Nhập số nguyên dương: "))

for i in range (1, a+1):
  if i**2 == a:
    print(f"{a} là số chính phương")
    break
  if i**2 > a:
    print(f"{a} không phải số chính phương")
    break
```

Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không.
Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28…

```python
a = int(input())
uoc = []

for i in range (1, a):
  if a%i == 0: uoc.append(i)

if a == sum(uoc): print(f"{a} là số hoàn hảo")
else: print(f"{a} không phải số hoàn hảo")
```

Bài 16: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu

Bài 18: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay
không

```python
n = int(input())

for k in range (0, n):
  if n == 3**k:
    print(f"{n} có dạng 3^{k}")
    break
  elif 3**k > n:
    print(f"{n} không có dạng 3^k")
    break
```

Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố.

```python
n = int(input())
i = 0

while n != 0:
  m = n%10
  def prime(m):
    if m<2:
      return False
    for i in range (2, int(m**0.5)+1):
      if m%i == 0:
        return False
    return True

  if prime(m) == True:
    i+=1
  n = (n-n%10)/10

print(i)
```
