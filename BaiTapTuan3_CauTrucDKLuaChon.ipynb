**W3A11**
[Phân loại tam giác]\
Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm tra 3 số đã nhập có tạo thành một tam giác hợp
lệ hay không. Nếu không in ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều", "Tam giác cân", hay
"Tam giác thường".

```python
sides = sorted(map(int, input().split()))
a, b, c = sides
if a + b <= c:
  print("Không phải tam giác")
if len(set(sides)) == 2:
  print("Tam giác cân")
elif len(set(sides)) == 1:
  print("Tam giác đều")
else:
  print("Tam giác thường")
```

**W3A14:**[Giải phương trình bậc nhất]\
Nhập vào hai số thực a và b của phương trình ax + b = 0. Tìm và in ra
nghiệm của phương trình, kết quả làm tròn đến số thập phân thứ 2. In ra "Vô
nghiệm" hoặc "Vô số nghiệm" trong trường hợp tương ứng.


```python
# :g để bỏ các số 0 thừa ở tận cùng dấu phẩy

a, b = map(float, input("nhập hệ số cho phương trình ax + b = 0: ").split())

if a == 0: print("phương trình vô nghiệm")
elif b == 0: print("phương trình vô số nghiệm")
else:
  print(f"""
  {a:g}x + {b:g} = 0
  x = {-b:g}/{a:g}
  x = {-b/a:.2f}""")
```

**W3A15:** [Xếp loại học lực] \
Nhập vào điểm trung bình của một học sinh và in ra học lực của học sinh đó.
Xếp loại học lực theo quy tắc:\
(>=) 8.0: Giỏi \
(>= 6.5): Khá \
(>= 5.0): Trung bình \
(< 5.0): Yếu

```python
while 1:
  try:
    n = float(input("Nhập điểm "))
    if n < 0 or n > 10:
      raise ValueError
    break
  except ValueError: print("Lỗi giá trị. Nhập lại")

if n >= 8: print("Giỏi")
elif 6.5<=n<8: print("Khá")
elif 5.0<=n<6.5: print("Trung bình")
else: print("Yếu")
```

**W3A16** [Rounding] \
Nhập vào một số thực: in ra kết quả làm tròn lên, tròn xuống, làm tròn tới số
nguyên gần nhất (không được sử dụng hàm sẵn có của pyton). Các kết quả
cách nhau một dấu cách

```python
n = float(input("nhập số thực: "))
a = n//1     # làm tròn xuống
b = n//1 + 1 # làm tròn lên

print(f"""
Kết quả làm tròn xuống: {a:g}
Kết quả làm tròn lên: {b:g}""")

if abs(n-a) > abs(n-b): print(f"Số nguyên gần nhất: {b:g}")
elif abs(n-a) < abs(n-b): print(f"Số nguyên gần nhất: {a:g}")
```
