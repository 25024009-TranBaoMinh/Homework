**Bài 3:** Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là
độ dài ba cạnh hay không, nếu  không thì in  ra màn  hình  'Khong la ba  canh
cua mot tam  giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân,
Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn
hình.

```python
a, b, c = sorted(map(float, input("nhập độ dài 3 cạnh: ").split()))
canh = a, b, c

P = (a+b+c)/2
S = (P * (P - a)*(P - b)*(P - c))**0.5

if a+b<=c: print("Không là ba cạnh của một tam giác.")
else:
  if len(set(canh)) == 1: print("Tam giác đều")
  elif len(set(canh)) == 2: print("Tam giác cân")
  elif a**2 + b**2 == c**2: print("Tam giác vuông")
  else: print("Tam giác thường")
  print(f"""Diện tích tam giác là: {round(S, 2)}\nChu vi tam giác là: {round(2*P, 2)}""")
```

**Bài 10:** Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một
điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
while 1:
  try:
    x, y = map(float, input("nhập tọa độ tâm đường tròn (x; y): ").split())
    r = float(input("nhập độ dài bán kính: "))
    x0, y0 = map(float, input("nhập tọa độ điểm A(x0; y0): ").split())
    break
  except ValueError: print("Học hành thế à?")

if ((x0-x)**2 + (y0-y)**2)**0.5 > r:
  print(f"Điểm A({x0:g}; {y0:g}) thuộc đường tròn tâm ({x:g}; {y:g}) bán kính {r:g}")
else: print(f"Điểm A({x0:g}; {y0:g}) không thuộc đường tròn tâm ({x:g}; {y:g}) bán kính {r:g}")
```
