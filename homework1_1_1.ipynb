9. Chuyển đổi số thành chữ số: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số
tương ứng (0 -> "Không", 1 -> "Một", ...) – Xem trước match …case

```python
n = int(input())
def chuyen(n):
  match n:
    case 0: return "Không"
    case 1: return "Một"
    case 2: return "Hai"
    case 3: return "Ba"
    case 4: return "Bốn"
    case 5: return "Năm"
    case 6: return "Sáu"
    case 7: return "Bảy"
    case 8: return "Tám"
    case 9: return "Chín"
print(chuyen(n))
```
