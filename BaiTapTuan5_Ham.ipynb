**W5A3**: Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def prime(n):

  if n<2: return False
  for i in range (2, int(n**0.5)+1):
    if n%i==0: return False

  return True

n = int(input())
print(prime(n))
```

**W5A4**: Viết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên.

```python
def perfect(n):

  sum = 0
  for i in range (1, n):
    if n%i==0: sum += i

  if n==sum: return True

  return False

n = int(input())
print(perfect(n))
```

**W5A6**: Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!)

```python
def facto(n):

  facto = 1
  for i in range (1, n+1):
    facto *= i

  return facto

n = int(input())
print(facto(n))
```

**W5A7**: Viết hàm giả lập một máy tính bỏ túi. Đầu vào: Hai số thực và phép toán cần thực hiện (+-*/). Đầu ra: Kết quả phép toán. Làm tròn chữ số thập phan thứ 2

```python
def calc(a, b, c):

  if c == "+": return round(a+b, 2)
  elif c == "-": return round(a-b, 2)
  elif c == "*": return round(a*b, 2)
  elif c == "/": return round(a/b, 2)

a, b = map(float, input().split())
c = str(input())
print(calc(a, b, c))
```

**W5A8**: Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này.

```python
def hamming(x, y):
  while len(str(x)) != len(str(y)):
    x, y = map(int, input().split())

  x = list(str(x))
  y = list(str(y))
  ham = 0

  for i in range (0, len(x)):
    if x[i] != y[i] : ham += 1

  return ham

x, y = map(int, input().split())
print(hamming(x, y))
```

**W5A9**: Viết hàm trả về tổng các chữ số của số nguyên dương n

```python
def sum(n):

  sum = 0
  while n != 0:
    sum += n%10
    n = (n-n%10)//10

  return sum
n = int(input())
print (sum(n))
```

**W5A10**: Viết hàm trả về true nếu hai từ a và b là đẳng cấu, ngược lại false.

```python
def dang_cau(a, b):
  a = list(a)
  b = list(b)

  if len(a) != len(b): return False

  for i in range (0, len(a)):
    for k in range (0, i):
      if a[i] == a[k]:
        if b[i] != b[k]: return False
      if b[i] == b[k]:
        if a[i] != a[k]: return False

  return True


a = str(input())
b = str(input())
print(dang_cau(a, b))
```

**W5A12**: Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và 0 <= |r| < |divisor).

```python
def divide_with_remainder(dividend, divisor):
  q = dividend // divisor
  r = dividend % divisor

  return f"({q:g}, {r:g})"

dividend, divisor = map(float, input().split())
print(divide_with_remainder(dividend, divisor))
```

**W5A15**: Viết hàm is_right_triangle(a, b, c) trả True về nếu ba số có thể là ba cạnh của tam giác vuông.

```python
def  is_right_triangle(a, b, c):
  a, b, c = sorted((a, b, c))
  if a+b>c and a+c>b and b+c>a and a**2 + b**2 == c**2:
    return True

  return False

a, b, c = map(float, input().split())
print(is_right_triangle(a, b, c))
```
