**W4A1**[SumOfNumBers] \
Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để
tính tống các số nguyên từ 1 đến n. (n <= 1000)

```python
def tinh_tong(n):
  tong = 0
  for i in range (1, n+1):
    tong += i

  return tong

n = int(input())
print(tinh_tong(n))
```

[PrimeNumber] \

Kiểm tra số này có phải là số nguyên tố hay không?

```python
def is_prime(n):
  for i in range (2, int(n**0.5)+1):
    if n%i==0:
      return False
  if n<2:
    return False
  return True

n = int(input())
while n <=0: n = int(input())
print(is_prime(n))
```

**W4A6**
[SumPrimesInRange] \
Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn
[a, b]. Mỗi dòng là một bộ test (a b)."

```python
def sum_prime(a, b):
  sum = 0
  for i in range (a, b+1):
    if i < 2: continue
    for k in range (2, int(i**0.5)+1):
      if i%k == 0:
        break
    else: sum += i

  return sum

a, b = sorted(map(int, input().split()))
print(sum_prime(a, b))
```

**W4A7**
[LargestPrimeFactor] \
Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n.

```python
def big_prime(n):
  for i in range (n, 0, -1):
    if n%i == 0:
      for k in range (2, int(i**0.5)+1):
        if i%k == 0: break
      else: return i

n = int(input())
print(big_prime(n))
```

**W4A9**[PerfectSquaresDistinctDigits] \
Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số
đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu
cách). Nếu không có, in ra (no number).

```python
def chinh_phuong(n):
  i = 0
  k = i**2
  so = []
  while k < n:
    k = i**2
    i += 1
    if len(set(str(k))) == len(str(k)) and k < n :
      so.append(k)

  return ' '.join(map(str, so))
n = int(input())
print(chinh_phuong(n))

#luôn có số 0 là chính phương => ko tồn tại TH no number
```

**W4A10**[LongestCollatzUpToN] \
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
nhỏ nhất.

```python
def collatz(n):
  lst = 1
  num = 0

  for i in range (1, n+1):
    lst_1 = 1                          # đặt lst_1 là độ dài tạm thời, dùng để so sánh với dãy dài nhất trước đó
    i_1 = i
    if i == 1:
      lst = 1
      num = 1
    else:
      while i_1 != 1:                     # thử từng số từ 1 đến n, mỗi lần dãy tăng 1 số sẽ dc lst đếm lại
        if i_1 % 2 == 0:
          i_1 /= 2
        else:
          i_1 = 3*i_1+1

        lst_1 += 1

      if lst_1 > lst:                  # nếu dãy sau lớn hơn dãy trước: cập nhập i và độ dài dãy cố định
        lst = lst_1
        num = i
      if lst_1 == lst and i < num:       # nếu dãy sau bằng dãy i: cập nhật i nhỏ hơn
        lst = lst_1
        num = i

  return (num, lst)

n = int(input())
num, lst = collatz(n)
print(num, lst)
```

[EvenDivisors] \
Viết chương trình đếm số lượng ước số chẵn của số nguyên dương n với n < 10^6

```python
def count_uoc(n):
  k = 0
  for i in range (1, n+1):
    if n%i == 0 and i%2 == 0:
      k += 1

  return k

n = int(input())
print(count_uoc(n))
```

**W4A13**
[ClosePairNumber] \
Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật:
Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại.
Ví dụ 220 và 284 là một cặp số thân thiết
220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55,
110. Tổng của 11 ước này cho chúng ta 284
284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng
vừa đúng bằng 220 \
Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không.
Nếu có, in ra màn hình "true", ngược lại, in ra "false"

```python
def than_thiet(n1, n2):
  sum1 = 0
  sum2 = 0

  for i in range (1, n1):
    if n1 % i == 0: sum1 += i
  for i in range (1, n2):
    if n2 % i == 0: sum2 += i

  if sum1 == n2 and sum2 == n1:
    return True

  return False

n1, n2 = map(int, input().split())
print(than_thiet(n1, n2))
```

**W4A14**
[GCD] \
Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình
tìm ước chung lớn nhất của hai số m và n.

```python
def uoc_chung(m, n):
  for i in range (m, 0, -1):
    if m % i == n % i == 0:
      return i
      break

m, n = sorted(map(int, input().split()))
print(uoc_chung(m, n))
```

**W4A15**
[ChickenAndDog] \
"Vừa gà vừa chó,
Bó lại cho tròn,
Ba mươi sáu con,
Một trăm chân chẵn"
Từ bài toán dân gian trên, mở rộng thành chương trình nhận đầu vào là tổng số
con và tổng số chân của gà và chó. Nếu tìm được số phù hợp, in ra số lượng gà và
số lượng chó. Ngược lại, in ra "invalid".

```python
def ga_cho(sum, sum_leg):
  C = (sum_leg - 2*sum) / 2
  G = sum - C
  if int(C) == C and int(G) == G:
    return (int(G), int(C))

  return "invalid"

sum, sum_leg = map(int, input().split())
G, C = ga_cho(sum, sum_leg)
print(G, C)
```
