**W4A18**
Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b

```python
def uoc_chung(a, b):
  uoc_chung = []
  for i in range (1, a+1):
    if a % i == 0 and b % i == 0:
      uoc_chung.append(i)

  return (uoc_chung)

a, b = sorted(map(int, input().split()))
print(*uoc_chung(a, b))
```

**W4A20**Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không

```python
def ktra(n):
  for i in range (0, n):
    if 2**i == n:
      return True

  return False

n = int(input())
print(ktra(n))
```

**W4A21**
Nhập vào số nguyên dương n, tính tổng các chữ số của n

```python
def sum(n):
  sum = 0
  while n != 0:
    n1 = n % 10
    sum += n1
    n = (n-n%10)//10

  return sum

n = int(input())
print(sum(n))
```

**W4A22**
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
def count(n):
  even = []
  odd = []
  while n != 0:
    n1 = n%10
    if n1 % 2 == 0: even.append(n1)
    else: odd.append(n1)
    n = (n - n%10)//10

  return (len(even), len(odd))

n = int(input())
even, odd = count(n)
print(even, odd)
```

**W4A24**
Nhập vào A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
def min(A):
  n = 1
  S = 0
  while S <= A:
    S += (1/n)
    n += 1

  return n, round(S, 4)

A = float(input())
n, S = min(A)
print(n, S)
```

**W4A25**
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
def min_max(lst):
  min = max = lst
  try:
    while lst != -1:
      lst = int(input())
      if lst > max:
        max = lst
      elif lst < min:
        min = lst
  except ValueError: pass

  return min, max

lst = int(input())
min, max = min_max(lst)
print(min, max)
```

**W4A26** [Fibonaci1] \
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A

```python
def fib(A):
  fib1 = fib2 = 1
  while fib2 <= A:
    fib1, fib2 = fib2, fib2 + fib1

  return fib1
A = int(input())
print(fib(A))
```

**W4A28**

Dùng cách: for trên index; KHÔNG dùng hàm split()

```python
def first(lst):
  first = []
  for j in range (0, len(lst)):
    if  lst[j] != " ":
      if lst[j+1] != " " or j == 0:
        first.append(lst[j])
      elif lst[j+1] == " ":
        first.append(lst[j])
        break

  return first
lst = str(input())
print(*first(lst), sep = "")
```

**W4A29**
Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành
thường và thường thành hoa.

```python
def convert(lst):
  converted = ""
  for i in lst:
    if i.isupper():
      converted += i.lower()
    elif i.islower():
      converted += i.upper()
    elif i == " ":
      converted += i

  return converted

lst = str(input())
print(convert(lst))
```

**W4A31**
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in
thường, bao nhiêu ký tự số

```python
def count(lst):
  num = let = 0
  for i in lst:
    if i.isalpha(): let += 1
    elif i.isdigit(): num += 1

  return num, let

lst = list(input())
num, let = count(lst)
print(count(lst))
```
